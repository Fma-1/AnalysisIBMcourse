## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
#after many try i decided to remove outliers of WorkWeekHrs


iqr = df['WorkWeekHrs'].quantile(0.75) - df['WorkWeekHrs'].quantile(0.25) 

upperBounds = df['WorkWeekHrs'].quantile(0.75) + (1.5 * iqr)
lowerBounds = df['WorkWeekHrs'].quantile(0.25) - (1.5 * iqr)


print(upperBounds, lowerBounds)
print(iqr)

df_new = df[(df['WorkWeekHrs'] <= upperBounds) & (df['WorkWeekHrs'] >= lowerBounds)]
df_new['WorkWeekHrs'].value_counts()
```

```python
df_new.plot(kind='scatter',
       x='Age',
       y='WorkWeekHrs',
       figsize=(14, 8)
       )

plt.title('Average hours worked at week by Age ')
plt.xlabel('Age')
plt.ylabel('hours/week')

plt.yticks(np.arange(35, df_new['WorkWeekHrs'].max(), 1, dtype=int))

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

sns.scatterplot(data=df_new, y="WorkWeekHrs", x="CodeRevHrs", size="Age", legend= False, sizes=(1, 2000), alpha = 0.30)

plt.title('Work Week Hours and Code Revision Hours by Age')
plt.ylabel('Average hours worked at week')
plt.xlabel('Average Code review hours at week')


plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, count(*) num
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY 2 DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']

df.plot(kind='pie',
        y='num',
        figsize=(9, 9),
        autopct='%1.1f%%',
        labels=None,
        pctdistance=1.12,
        colors=colors_list
        )
plt.legend(labels = df['DatabaseDesireNextYear'])
plt.title('Top 5 Databases to Learn next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 and 35 
"""
df = pd.read_sql_query(QUERY,conn)

df_grp = df.groupby(['Age'], as_index=True).median()

df_grp


df_grp[['WorkWeekHrs','CodeRevHrs']].plot(kind='area',
            alpha= 0.35,
            figsize=(20, 10))
plt.title('Stacked chart of median Work Week and Code Revision Hours per week for the age group 30 to 35.')
plt.ylabel('Hours')
plt.xlabel('Age')
plt.show() 
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
AND ConvertedComp IS NOT NULL
"""
df = pd.read_sql_query(QUERY,conn)

df_grp = df.groupby(['Age'], as_index=False).median()

df_grp.plot(kind='line',
       x='Age',
       y='ConvertedComp',
       figsize=(14, 8),
       legend=None
       )

plt.title('Median Annual Salary by Age ')
plt.xlabel('Age')
plt.ylabel('Annual Salary')
plt.legend()

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, count(*) num
FROM master
GROUP BY MainBranch
ORDER BY 1 DESC
"""
df = pd.read_sql_query(QUERY,conn)

df.set_index('MainBranch', inplace=True)

ax = df.plot(kind='barh',
       figsize=(14, 4)
       )
lg = ax.get_legend()
lg.remove()
plt.ylabel('')

plt.annotate(df.iloc[0:1].values[0], xy=(300 ,- 0.05), color='white')
plt.annotate(df.iloc[1:2].values[0], xy=(10000, 0.95), color='white')

plt.show()


```
